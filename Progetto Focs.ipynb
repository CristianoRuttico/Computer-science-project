```python
#pulizia di '='
def pulizia1(row):
    compi = re.compile('(^\=?)')
    subs = compi.sub('', row['world_rank'])
    
    if subs:
        return subs
        
    else:
        return row['world_rank'] 
        
    
```

```python
#funz pulizia e media campo world_rank di shangai e times
def pulizia_e_mean(row):
    compilato = re.compile('(\d+)(-)(\d+)')
    occorrenza = compilato.search(row['world_rank'])
    if occorrenza:
        return (int(occorrenza.group(1))+ int(occorrenza.group(3)))/2
    else:    
        return int(row['world_rank'])
        
```

## 5. Consider only the most recent data point of the times dataset. Compute the number of male and female students for each country.


```python
#modifico la colonna di interesse
times.rename(columns ={'female_male_ratio':'FemmSuMasch'}, inplace =True)
times
```

```python
#piccolo check per vedere se posso usare regex su rapporto femmine/maschi
times.dtypes
```

```python
#prendo i più recenti raggruppando per university 
timesrecentigruppati = times.loc[times.groupby('university')['year'].idxmax()]
timesrecentigruppati
```

```python
#visualizzo i valori del rapporto maschi femmine e noto che ci sono dei '-' e molti nan 
timesrecentigruppati['FemmSuMasch'].values
```

```python
#converto i '-' in nan
timesrecentigruppati['FemmSuMasch'].replace('-', np.nan, inplace=True)
```

```python
#tolgo i nan e controlloche tutto sia andato a buon fine
timesrecentigruppati.dropna(subset=['num_students', 'FemmSuMasch'], inplace =True)
timesrecentigruppati['FemmSuMasch'].values
```

```python
#trovo il numero delle femmine
def numerofemmine(row):
    pat = re.compile('(\d+)\s+(\:)\s+(\d+)')
    okko = pat.search(row['FemmSuMasch'])
    g1 = okko.group(1)
    g2 = okko.group(2)
    a = float(row['num_students'] * int(g1) /100) *1000
    return round(a)
```

```python
#applico la funzione e mostro il numero di femmine
timesrecentigruppati['Femmine'] = timesrecentigruppati.apply(numerofemmine, axis =1)
timesrecentigruppati
```

```python
#trovo numero dei maschi
def numeromaschi(row):
    return round((row['num_students']*1000) - row['Femmine'])           
```

```python
#creo la colonna maschi
timesrecentigruppati['Maschi'] = timesrecentigruppati.apply(numeromaschi, axis=1)
```

```python
#mostro il numero dei maschi
timesrecentigruppati
```

```python
#mostro il numero di maschi e femmine per ogni nazione
timesrecentigruppati.groupby('country', as_index =False).sum()[['country','Maschi', 'Femmine']]
```
